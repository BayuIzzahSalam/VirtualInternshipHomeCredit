```python
df['DATE_BIRTH'] = pd.to_datetime(df['DATE_BIRTH'])
df.info()
```

```python
from datetime import date
```

```python
yearBirth = []
for index, row in df.iterrows():
    yearBirth.append(row['DATE_BIRTH'].year)
```

```python
age = []
for x in yearBirth :
    string = str(x)
    integer = int(string)
    age1 = date.today().year - integer
    age.append(age1)
```

```python
print(age)
```

```python
lateTen = []
for i in age:
    if i < 20:
        lateTen.append(i)
        
print(sum(lateTen))
```

```python
twenty = []
for i in age:
    if i > 20 and i < 30:
        twenty.append(i)
        
print(len(twenty))
```
